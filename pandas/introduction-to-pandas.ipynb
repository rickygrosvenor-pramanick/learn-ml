```python
import pandas as pd
```

```python
# 2 main pandas datatypes - 1st one is Series
f1_series = pd.Series(["Red Bull Racing RBPT", "Mclaren Mercedes", "Scuderia Ferrari"])
wins_series = pd.Series([7 , 3, 1])
```

```python
f1_series
```

```python
# As assessed a series is essentially an array extrapolated to a table where the index is the first column
# and the second column is the actual data.

# As a result, we can call series 1-dimensional as it only has 1 column of data.
```

```python
# 2 main pandas datatypes - 2nd one is DataFrames
f1_df = pd.DataFrame({"Team" : f1_series, "Wins" : wins_series})
f1_df
```
