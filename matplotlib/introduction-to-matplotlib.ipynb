# A matplotlib workflow

```python
# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("../simple-plot.png")
```

## Making Figures with NumPy Arrays

```python
import numpy as np
x = np.linspace(0, 10, 100) # Returns num evenly spaced intervals between start and stop
x
```

```python
# Plot the Data and create a line plot (default)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x**2)
```

```python
# Using the same data, plot a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Using the same data, plot a bar plot
fig, ax = plt.subplots()
ax.bar(x, np.sin(x))
```

```python
# Making a plot from a dictionary
nut_butter_prices = {"Peanut" : 8.5,
                    "Almond" : 10,
                    "Macadamia" : 20}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Prices", xlabel="Nut", ylabel="Price ()")
```

```python
# Horizontal Bar Plots
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some Data for Histograms and Subplots
# Return a sample (or samples) from the "standard normal" distribution.
data = np.random.randn(1000)
data
```

```python
fig, ax = plt.subplots()
ax.hist(data, bins=100)
```

## Two Options for Subplots

```python
# Subplot Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(data, np.sinh(data))
ax2.scatter(data, np.exp(data))
ax3.bar(data, data/2)
ax4.hist(data, bins=50)
```

```python
# Subplot Option 2 (Preferred)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot on different axis
ax[0][0].plot(data, np.sinh(data))
ax[0][1].scatter(data, np.exp(data))
ax[1][0].bar(data, data/2)
ax[1][1].hist(data, bins=50)
```
