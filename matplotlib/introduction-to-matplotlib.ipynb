# A matplotlib workflow

```python
# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("../simple-plot.png")
```

## Making Figures with NumPy Arrays

```python
import numpy as np
x = np.linspace(0, 10, 100) # Returns num evenly spaced intervals between start and stop
x
```

```python
# Plot the Data and create a line plot (default)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x**2)
```

```python
# Using the same data, plot a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Using the same data, plot a bar plot
fig, ax = plt.subplots()
ax.bar(x, np.sin(x))
```

```python
# Making a plot from a dictionary
nut_butter_prices = {"Peanut" : 8.5,
                    "Almond" : 10,
                    "Macadamia" : 20}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Prices", xlabel="Nut", ylabel="Price ()")
```

```python
# Horizontal Bar Plots
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

## Two Options for Subplots

```python
# Subplot Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(data, np.sinh(data))
ax2.scatter(data, np.exp(data))
ax3.bar(data, data/2)
ax4.hist(data, bins=50)
```

```python
# Subplot Option 2 (Preferred)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot on different axis
ax[0][0].plot(data, np.sinh(data))
ax[0][1].scatter(data, np.exp(data))
ax[1][0].bar(data, data/2)
ax[1][1].hist(data, bins=50)
```

# 4. Plotting with pandas using the OO method

For more complicated plots, you'll want to use the OO method.

```python
# Perform data analysis on patients over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
# Pyplot Method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c='target')
```

```python
# Object Oriented Method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)

# To adjust the range of x-values in the x-axis, use the set_xlim() function
# ax.set_xlim([45, 100])
```

```python
# Object Oriented Method FROM SCRATCH
fig, ax = plt.subplots(figsize=(10,6))

# Plot the Data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"], 
                     c=over_50["target"])

# Customise the Plot
ax.set(title="Effect of Age on Cholestorel on Over 50s", xlabel="Age", ylabel="Cholestorol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(y=np.mean(over_50["chol"]), linestyle='--', c="red")
```

```python
over_50
```

```python
# Subplots of age, chol, thalach

fig, ax = plt.subplots(figsize=(10,10), nrows=2)

chol_ax = ax[0].scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
thalach_ax = ax[1].scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])

ax[0].set(title="Effect of Age on Cholestorel on Over 50s", xlabel="Age", ylabel="Cholestorol")
ax[1].set(title="Effect of Age on Max Heart Rate on Over 50s", xlabel="Age", ylabel="Max Heart Rate")

ax[0].legend(*chol_ax.legend_elements(), title="Targets")
ax[1].legend(*chol_ax.legend_elements(), title="Targets")

ax[0].axhline(y=np.mean(over_50["chol"]), linestyle='--', c='red')
ax[1].axhline(y=np.mean(over_50["thalach"]), linestyle='--', c='blue')

# Add a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold')
```
