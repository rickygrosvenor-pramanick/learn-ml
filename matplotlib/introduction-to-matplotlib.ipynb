# Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot()
```

```python
# To remove the [], use semi-colon at the end of the line plt.plot(); or plt.show after the plt.plot
plt.plot()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
plt.show()
```

```python
# Creating a plot with the OO verison, confusing way first
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# Confusing #2
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# Easier and more robust going forward (what we're going to use)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```
