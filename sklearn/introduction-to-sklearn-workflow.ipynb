# Introduction to Scikit-Learn

## Typical Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1. Get the Data Ready
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Scikit Learn Typical Variables
# X - Features Matrix
# y - Target Feature
```

```python
# Create X (Features Matrix)
X = heart_disease.drop("target", axis=1)

# Create y (Labels)
y = heart_disease["target"]
```

```python
# 2. Choose the Right Model and Hyperparameters
# Notice this is a binary classification problem - so we'll use a classification model such as Random Forest or Logistic Regression
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

# 80% of the data is used for training and 20% used for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
clf.fit(X_train, y_train)
```

```python
X_test
```

```python
# Make a Prediction
y_preds = clf.predict(X_test)
```

```python
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the models on the training data and the test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a Model i.e try different values for hyperparameters - in this case n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}")
    print(" ")
```

```python
# From this, it seems the optimal number of estimators is 40
clf = RandomForestClassifier(n_estimators=40).fit(X_train, y_train)
```

```python
# 6. Save a model and load it
import pickle
pickle.dump(clf, open("random_forst_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forst_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```
